```python
class User:
  def __init__(self, id, name, username, email):
    self.id = id
    self.name = name
    self.username = username
    self.email = email
```

```python
user1 = User(1,"Engin", "ed", "engin@kodlama.io")
user1.name
```

```python
class Post:
  def __init__(self, id, title, body, user):
    self.id = id
    self.title = title
    self.body = body
    self.user = user
```

```python
class Comment:
  def __init__(self,id,name,email,body,post):
    self.id = id
    self.name = name
    self.email = email
    self.body = body
    self.post = post
```

```python
user1 = User(1, "Engin", "ed","enginqgmail")
post1 = Post(1, "Python", "Classes", user1)
post2 = Post(1, "Python", "Classes", user1)
comment1 = Comment(1, "yorumum", "engin@", "çok güzel", post2)

comment1.post.user.name
```
