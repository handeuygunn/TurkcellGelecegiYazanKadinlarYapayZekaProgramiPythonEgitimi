```python
orders = [ {"customer": "Ali", "items": [{"name": "Laptop", "price": 15000}, {"name": "Mouse", "price": 500}]},
 {"customer": "Ayşe", "items": [{"name": "Phone", "price": 12000}, {"name": "Headphones", "price": 2000}]},
 {"customer": "Mehmet", "items": [{"name": "Tablet", "price": 7000}, {"name": "Keyboard", "price": 1000}, {"name": "Mouse", "price": 600}]},
 {"customer": "Zeynep", "items": [{"name": "Monitor", "price": 8000}, {"name": "Phone", "price": 11000}]}, ]
```

```python
#Toplam sipariş tutarı en yüksek olan müştesi
maxPrice = 0
maxCustomer = ""
for order in orders:
  items = order["items"]
  totalPrice = sum(item["price"] for item in items)
  if totalPrice > maxPrice:
    maxPrice = totalPrice
    maxCustomer = order["customer"]
print(f"Toplam sipariş tutarı en yüksek olan müşteri: {maxCustomer}")
```

```python
#Her müşterinin toplam sipariş tutarını içeren bir sözlük oluşturun.
TotalOrder = {}
for order in orders:
  items = order["items"]
  totalPrice = sum(item["price"] for item in items)
  TotalOrder[order["customer"]] = totalPrice
print(f"Her müşterinin toplam sipariş tutarı: {TotalOrder}")
```

```python
#En çok satılan ürünü bulun. (Birden fazla müşteri tarafından sipariş edilen ürün)
products = {}
for order in orders:
  items = order["items"]
  for item in items:
    if item["name"] in products:
      products[item["name"]] += 1
    else:
      products.update({item["name"]:+1})
for product in products:
  if products[product] == max(products.values()):
    print(f"En çok satılan ürün: {product}")
```

```python
#Ortalama sipariş tutarını hesaplayın.
TotalOrder["price"] = sum(TotalOrder.values())/len(TotalOrder)
print(f"Ortalama sipariş tutarı: {TotalOrder['price']}")
```

```python
#Siparişlerinde toplam tutarı 10.000 TL’nin üzerinde olan müşterileri listeleyin.
plusTen = []
for order in TotalOrder:
  if TotalOrder[order] > 10000:
    plusTen.append(order)
print(f"Toplam tutarı 10.000 TL'nin üzerinde olan müşteriler: {plusTen}")
```

```python
employees = [ {"name": "Ahmet", "department": "IT", "projects": [{"name": "CRM", "completed": True, "hours": 120}, {"name": "Security", "completed": False, "hours": 80}]},
{"name": "Elif", "department": "HR", "projects": [{"name": "Hiring System", "completed": True, "hours": 90}, {"name": "Employee Portal", "completed": True, "hours": 100}]},
{"name": "Mehmet", "department": "IT", "projects": [{"name": "Cloud Migration", "completed": True, "hours": 150}, {"name": "Data Analysis", "completed": True, "hours": 130}]},
{"name": "Ayşe", "department": "Marketing", "projects": [{"name": "Social Media", "completed": True, "hours": 60}, {"name": "SEO", "completed": False, "hours": 50}]} ]
```

```python
#En çok saat harcayan çalışanı bulun.
totalHoursDict = {}
for employee in employees:
  totalHours = sum(project["hours"] for project in employee["projects"])
  totalHoursDict[employee["name"]] = totalHours
for employee in totalHoursDict:
  if totalHoursDict[employee] == max(totalHoursDict.values()):
    print(f"En çok saat harcayan çalışan: {employee}")
```

```python
#Tamamlanan proje sayısı en yüksek olan çalışanı belirleyin.
completedProjectsDict = {}
for employee in employees:
  completedProjects = sum(1 for project in employee["projects"] if project["completed"])
  completedProjectsDict[employee["name"]] = completedProjects
for employee in completedProjectsDict:
  if completedProjectsDict[employee] == max(completedProjectsDict.values()):
    print(f"Tamamlanan proje sayısı en yüksek olan çalışan: {employee}")
```

```python
#Her departmandaki toplam çalışma saatlerini hesaplayın.**
totalHoursDict = {}
for employee in employees:
  totalHours = sum(project["hours"] for project in employee["projects"])
  totalHoursDict[employee["department"]] = totalHours
print(f"Her departmandaki toplam çalışma saatleri: {totalHoursDict}")
```

```python
#En çok proje yapan departmanı bulun.**
totalProjectsDict = {}
for employee in employees:
  totalProjects = sum(1 for project in employee["projects"])
  totalProjectsDict[employee["department"]] = totalProjects
print(f"En çok proje yapan departman: {max(totalProjectsDict, key=totalProjectsDict.get)}")
print(totalProjectsDict)
```

```python
#Tüm projelerin tamamlanma yüzdesini hesaplayın.
completedProjectsDict = {}
for employee in employees:
  totalProjects = sum(1 for project in employee["projects"])
  completedProjects = sum(1 for project in employee["projects"] if project["completed"])
  completedProjectsDict[employee["name"]] = completedProjects/totalProjects
print(f"Tüm projelerin tamamlanma yüzdesi: {completedProjectsDict}")
```

```python
stock = [ {"category": "Meyve", "items": [{"name": "Elma", "price": 10, "quantity": 50}, {"name": "Muz", "price": 15, "quantity": 30}]},
{"category": "Sebze", "items": [{"name": "Domates", "price": 8, "quantity": 60}, {"name": "Patates", "price": 5, "quantity": 100}]},
{"category": "İçecek", "items": [{"name": "Maden Suyu", "price": 3, "quantity": 150}, {"name": "Kola", "price": 12, "quantity": 80}]} ]
```

```python
#Stokta toplam kaç ürün olduğunu hesaplayın.**
for types in stock:
  total = sum(item["quantity"] for item in types["items"])
print(f"Stokta toplam {total} ürün bulunmaktadır.")
```

```python
#En değerli ürünü (price * quantity en yüksek olan) bulun.
valueDict = {}
for types in stock:
  for item in types["items"]:
    value = item["price"] * item["quantity"]
    valueDict[item["name"]] = value
for item in valueDict:
  if valueDict[item] == max(valueDict.values()):
    print(f"En değerli ürün: {item}")
```

```python
#Her kategori için toplam stok değerini hesaplayan bir sözlük oluşturun.**
cateDict = {}
for types in stock:
  total = sum(item["quantity"] for item in types["items"])
  cateDict[types["category"]] = total
print(f"Her kategori için toplam stok değerleri: {cateDict}")
```

```python
#Stok miktarı en düşük olan ürünü bulun.
stockDict = {}
for types in stock:
  for item in types["items"]:
    stockDict[item["name"]] = item["quantity"]
for item in stockDict:
  if stockDict[item] == min(stockDict.values()):
    print(f"Stok miktarı en düşük olan ürün: {item}")
```

```python
#Ortalama ürün fiyatını hesaplayın.**
totalPrice = sum(item["price"]*item["quantity"] for types in stock for item in types["items"])
print(f"Ortalama ürün fiyat: {totalPrice/total}")
```
