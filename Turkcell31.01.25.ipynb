```python
class MyMath:
  def __init__(self, n1, n2): #reserved name for constructors
    self.number1 = n1 # python self i parametre gibi görmüyor
    self.number2 = n2

  def sum(self):
    return self.number1 + self.number2

  def subtract(self):
    return self.number1 - self.number2

  def multiply(self):
    return self.number1 * self.number2

  def divide(self):
    return self.number1/ self.number2
```

```python
x = MyMath(2,3) # create instance
print(x.sum())
print(x.subtract())
print(x.multiply())
print(x.divide())
```

```python
class Product:
  def __init__(self,id, name, price, stock):
    print("class başladı")
    self.id = id
    self.name = name
    self.price = price
    self.stock = stock
```

```python
product1 = Product(1, "Elma", 20, 20)
product2 = Product(2, "Armut", 20, 20)
product3 = Product(3, "ÜZÜM", 20, 20)

products = [product1, product2, product3]

for product in products:
  print(product.name, product.id, product.price, product.stock)
```

```python
class Car:
  def __init__(self, brandName, modelName, price):
    print("Araba üretildi")
    self.brandName = brandName
    self.modelName = modelName
    self.price = price

  def sell(self):
    print("Araba satıldı", self.brandName,self.modelName, self.price)
```

```python
car1 = Car("BMW", "I4", 100)
car1.sell()
```
