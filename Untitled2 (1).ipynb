```python
#Bir e ticaret sistemindeki Product, Category, Cart, User, Order, OrderDetail,
```

```python
class Product:
  def __init__(self, name, price, stock, category):
    self.name = name
    self.price = price
    self.stock = stock
    self.category = category
```

```python
class Category:
  def __init__(self, name):
    self.name = name
```

```python
class Cart:
  def __init__(self, user, products):
    self.user = user
    self.products = products
```

```python
class User:
  def __init__(self, name, email, password):
    self.name = name
    self.email = email
    self.password = password
```

```python
class Order:
  def __init__(self, user, cart, order_date, order_detail):
    self.user = user
    self.cart = cart
    self.order_date = order_date
    self.order_detail = order_detail
```

```python
class OrderDetail:
  def __init__(self, product, quantity, price):
    self.product = product
    self.quantity = quantity
    self.price = price
```

```python
category1 = Category("Electronics")
category2 = Category("Clothing")
category1.name
```

```python
product1 = Product("Laptop", 1000, 10, category1)
product2 = Product("Phone", 500, 20, category1)
product3 = Product("Shirt", 20, 50, category2)
product4 = Product("Jeans", 30, 40, category2)
product1.category.name
```

```python
orderDetail1 = OrderDetail(product1, 2, 1000)
orderDetail1.product.name
```
